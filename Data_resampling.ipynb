```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('dNetflix_cleaned_trades.csv', parse_dates = [["DATE","TIME"]], index_col=0)
df.head()
```

```python
price_resampled = df['PRICE'].resample("5T").mean()
price_resampled= price_resampled.dropna()
price_resampled.to_csv('data_resampled.csv')
```

```python
plt.figure(figsize=(16,10))
plt.plot(price_resampled)
```
